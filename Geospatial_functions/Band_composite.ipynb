```python
import rasterio as rio
import numpy as np
from matplotlib import pyplot as plt
from rasterio.plot import show
```

```python
nir_band = rio.open('LC08_L1TP_140043_20180109_20180119_01_T1/LC08_L1TP_140043_20180109_20180119_01_T1_B5.TIF')
red_band = rio.open('LC08_L1TP_140043_20180109_20180119_01_T1/LC08_L1TP_140043_20180109_20180119_01_T1_B4.TIF')
green_band = rio.open('LC08_L1TP_140043_20180109_20180119_01_T1/LC08_L1TP_140043_20180109_20180119_01_T1_B3.TIF')
blue_band = rio.open('LC08_L1TP_140043_20180109_20180119_01_T1/LC08_L1TP_140043_20180109_20180119_01_T1_B2.TIF')
```

```python
nir = np.empty(shape = (nir_band.shape))
nir = nir_band.read(1).astype(float)

red = np.empty(shape = (red_band.shape))
red = red_band.read(1).astype(float)

green = np.empty(shape = (green_band.shape))
green = green_band.read(1).astype(float)

blue = np.empty(shape = (blue_band.shape))
blue = blue_band.read(1).astype(float)
```

```python
fig = plt.imshow(nir)
#fig.set_cmap('inferno')
plt.colorbar()
plt.axis('off') #Since numpy array won't show actual coordinates!
```

```python
from rasterio.plot import show
show(nir)
```

```python
fcc = np.dstack((nir, red, green))
fcc = fcc/fcc.max()             # or any other shrinking
plt.imshow(fcc)
plt.axis('off')
```

```python
tcc = np.dstack((red, green, blue))
tcc = tcc/tcc.max()
plt.imshow(tcc)
plt.axis('off')
```
